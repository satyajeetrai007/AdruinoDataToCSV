```python
import serial
import time
import csv

# Configure the serial port
port = 'COM13'  # Change this to your Arduino's serial port
baud_rate = 9600  # Match this to the Arduino baud rate
timeout = 1  # Timeout for read operations

# Open the serial port
ser = serial.Serial(port, baud_rate, timeout=timeout)

# Give the connection a second to settle
time.sleep(2)

# Open a CSV file to store the data
with open('sensor_data.csv', 'a', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
   
    # Optionally, write header only if the file is empty
    csvfile.seek(0, 2)  # Move the cursor to the end of the file
    if csvfile.tell() == 0:  # Check if file is empty
        csv_writer.writerow(['Sensor1', 'Sensor2'])  # Write header

    try:
        while True:
            if ser.in_waiting > 0:  # Check if there's data in the buffer
                line = ser.readline().decode('utf-8').rstrip()  # Read a line and decode
                print(line)  # Print to the console (optional)
               
                # Split the line into sensor values
                sensor1, sensor2 = line.split(',')
               
                # Convert to float or int if needed
                sensor1 = float(sensor1)  # or int(sensor1)
                sensor2 = float(sensor2)  # or int(sensor2)
               
                # Write to the CSV file
                csv_writer.writerow([sensor1, sensor2])
    except KeyboardInterrupt:
        print("Program stopped by user.")
    finally:
        ser.close()  # Close the serial port
        print("Serial port closed.")
```
